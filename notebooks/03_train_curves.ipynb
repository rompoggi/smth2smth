# 03 - Training Curves

**Goal:** compare training runs side-by-side.

Reads metrics emitted by `pipelines/train.py` (CSV/JSON logs and/or checkpoint metadata) and plots loss, top-1, and top-5 across epochs. Useful to compare:
- Track A vs Track B,
- pretrained vs from-scratch,
- different model variants.

Adapt the log path conventions to whatever the trainer in Phase 4 emits.

```python
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path

REPO_ROOT = Path.cwd().parent
SRC_DIR = REPO_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
RUNS_DIR = REPO_ROOT / "outputs"

def load_run_metrics(run_dir: Path) -> pd.DataFrame:
    """Placeholder: adapt to whatever the trainer logs (CSV, JSON, etc.)."""
    metrics_csv = run_dir / "metrics.csv"
    if not metrics_csv.exists():
        return pd.DataFrame()
    df = pd.read_csv(metrics_csv)
    df["run"] = run_dir.name
    return df

all_runs = [load_run_metrics(p) for p in sorted(RUNS_DIR.glob("*"))] if RUNS_DIR.exists() else []
df = pd.concat([d for d in all_runs if not d.empty], ignore_index=True) if all_runs else pd.DataFrame()
df.head()
```

```python
if not df.empty:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.lineplot(data=df, x="epoch", y="train_loss", hue="run", ax=axes[0])
    axes[0].set_title("Train loss")
    sns.lineplot(data=df, x="epoch", y="val_top1", hue="run", ax=axes[1])
    axes[1].set_title("Validation top-1")
    plt.tight_layout()
else:
    print("No run metrics yet. Train one model first.")
```
