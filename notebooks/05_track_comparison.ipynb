## Per-class delta (B - A)

Highlights classes where pretraining/open-world setup helps the most or hurts.

```python
# Replace these series with actual per-class accuracies once eval is wired up.
acc_a = pd.Series(dtype=float, name="track_a")
acc_b = pd.Series(dtype=float, name="track_b")

if not acc_a.empty and not acc_b.empty:
    delta = (acc_b - acc_a).sort_values()
    fig, ax = plt.subplots(figsize=(12, 4))
    delta.plot.bar(ax=ax)
    ax.set_title("Per-class accuracy delta (B - A)")
    ax.set_ylabel("delta")
    plt.tight_layout()
else:
    print("Per-class accuracies not yet available.")
```
